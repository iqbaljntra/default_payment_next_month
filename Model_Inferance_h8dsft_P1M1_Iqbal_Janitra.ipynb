# IMPORT LIBRARIES

```python
import pandas as pd
import numpy as np 
import pickle
import ast
import joblib
```

# LOAD MODEL

```python
with open('model_rf_model.pkl', 'rb') as file_1:
  model_rf_model= joblib.load(file_1)

with open('model_scaler.pkl', 'rb') as file_2:
  model_scaler=joblib.load(file_2)

with open('list_num_cols.txt', 'rb') as file_4:
  num_cols= joblib.load(file_4)

with open('list_cat_cols.txt', 'rb') as file_5:
  cat_cols= joblib.load(file_5)
```

# MAKE RANDOM DATA

```python
random = np.random.default_rng(seed=2895)
limit_balance = random.choice(np.arange(10000, 800000, 10000), size=30)
pay_0 = random.integers(-2.0, 8.0, size=30)
pay_2 = random.integers(-2.0, 7.0, size=30)
pay_3 = random.integers(-2.0, 7.0, size=30)
pay_4 = random.integers(-2.0, 8.0, size=30)
pay_5 = random.integers(-2.0, 7.0, size=30)
pay_6 = random.integers(-2.0, 7.0, size=30)
data_inf = pd.DataFrame({'limit_balance' : limit_balance, 'pay_0' : pay_0, 'pay_2' : pay_2, 'pay_3' : pay_3, 'pay_4' : pay_4, 'pay_5' : pay_5, 'pay_6' : pay_6})
data_inf
```

# MODEL

```python
data_inf_num = data_inf[num_cols]
data_inf_cat = data_inf[cat_cols]
```

```python
data_inf_num_scaled = model_scaler.transform(data_inf_num)
data_inf_cat_encoded = data_inf_cat
```

```python
data_inf_final = np.concatenate([data_inf_num_scaled, data_inf_cat], axis=1)
```
